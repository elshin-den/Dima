```python
import scipy.integrate as integrate
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(a, b, n1, n2):
    def g(t):
        return 1.0 / math.pow(1.0 - t, n1) / math.pow(a - b*t, n2)
    return g
```

```python
A = 1.0724
B = 1.7659
X = [0.560958904, 0.400684932, 0.280239521, 0.186826347, 0.14011976]
n1_linspace = np.linspace(-1, 2, 61)
n2_linspace = np.linspace(-1, 2, 61)
```

```python
def integral(x):
    for i, n1 in enumerate(n1_linspace):
        for j, n2 in enumerate(n2_linspace):
            print('x={0}, n1={1}, n2={2}, value={3}'.format(x, n1, n2, integrate.quad(f(A, B, n1, n2), 0, x)[0]))
```

```python
# plt.subplot(len(X), 1, 1)
for x in X:
    integral(x)
```
